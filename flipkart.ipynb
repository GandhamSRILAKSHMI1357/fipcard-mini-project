```python
import matplotlib.pyplot as plt
import seaborn as sns

pivot = df.pivot_table(values='rating', index='category', aggfunc='mean')

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, cmap="YlOrRd", linewidths=1)
plt.title("Average Product Rating by Category")
plt.show()
```

This cell creates a heatmap to show the average product rating for each category. The dataset is first grouped by category, and the mean rating is calculated. In the heatmap, each cell represents a category, with its color intensity indicating the average rating â€” darker shades represent higher ratings. The numeric value is also displayed in each cell, making it easy to compare the average ratings of different categories at a glance.


```python
plt.figure(figsize=(10,6))
plt.scatter(
    df['rating'],
    df['price'],
    s=df['review_count']/5,
    c=df['discount'],
    cmap='plasma',
    alpha=0.7
)

plt.colorbar(label="Discount (%)")
plt.xlabel("Rating")
plt.ylabel("Price")
plt.title("Bubble Chart: Rating vs Price vs Reviews & Discount")
plt.show()
```

This cell creates a bubble chart that visualizes multiple aspects of the products simultaneously. Each bubble represents a product, with its position on the x-axis showing the rating and on the y-axis showing the price. The size of the bubble represents the number of reviews, with larger bubbles indicating more popular products. The color of the bubble represents the discount percentage, with different colors showing varying discount levels. This plot allows analysis of how rating, price, popularity, and discount interact for different products.


```python
plt.figure(figsize=(10,6))
sns.violinplot(
    x='category',
    y='price',
    data=df,
    palette='Set3'
)
plt.xticks(rotation=30)
plt.title("Price Distribution by Category (Violin Plot)")
plt.show()
```

This cell creates a violin plot to show the distribution of product prices within each category. Each violin represents a category, and the width of the shape at different price levels indicates the density of products at that price. Wider sections show where more products are concentrated, while narrower sections show fewer products. This visualization helps understand the range, spread, and concentration of prices within each category.


```python
plt.figure(figsize=(10,6))
sns.stripplot(
    x='discount',
    y='price',
    data=df,
    jitter=True,
    palette='rainbow'
)
plt.title("Effect of Discount on Product Price")
plt.show()
```

This cell creates a strip plot to visualize the relationship between product discount and price. Each point represents a product, positioned according to its discount percentage on the x-axis and its price on the y-axis. The jitter spreads points slightly to prevent overlap, making clusters and patterns easier to see. This plot helps identify how prices vary across different discount levels and highlights any trends in pricing for discounted products.


```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['rating'],
    bins=10,
    kde=True
)

plt.title("Distribution of Product Ratings with Smooth Density")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.show()
```

This cell creates a histogram combined with a kernel density estimation (KDE) curve to show the distribution of product ratings. The histogram bars represent the number of products in each rating range, while the smooth KDE curve highlights the overall trend of the ratings. This visualization helps identify which ratings are most common and provides insight into overall customer satisfaction patterns.
